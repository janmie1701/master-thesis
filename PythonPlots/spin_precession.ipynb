```python
q = 3                                         
chi1 = [-.2, -.4, -.6]
chi2 = [.2, .4, .6]

# calculate reference frequency with my get_f_ref function
t_ref = -2000
f_ref = get_f_ref(q, chi1, chi2, t_ref) 

# get strain and spin dynamics
t, h, dyn = sur(q, chi1, chi2, dt=dt, f_low=f_low, f_ref=f_ref, precessing_opts={'return_dynamics': True})                             
```
