# (2,$\pm$2) modes of superkick configuration with NRSur7dq4

```python
# import matplotlib modules
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
from mpl_toolkits.axes_grid1.inset_locator import mark_inset

# import main code
import sys
sys.path.append('/home/jannik/MA/fertige-MA/gitrepo')
from master_project import *
```

```python
# configuration
q = 1
chi1 = [0.8,0.,0.]
chi2 = [-0.8,0.,0.]

# calculate waveform with NRSur7dq4 surrogate model
dt = 0.1        # timestep size (units of M)
f_low = 0       # initial frequency (f_low = 0 returns the full surrogate)
t, h, dyn = sur(q, chi1, chi2, dt=dt, f_low=f_low)

# get real part of inertial frame modes
h_iner_22 = np.real(h[2,2])
h_iner_2m2 = np.real(h[2,-2])
h_iner_21 = np.real(h[2,1])
h_iner_4m4 = np.real(h[4,-4])
```

```python
# setup plotting
fig = plt.figure()
ax1 = fig.add_subplot(111)
axins = ax1.inset_axes([0.08, 0.03, 0.43, 0.37])

# main plot
ax1.plot(t, h_iner_22, color='darkorange', label=r'$\ell=2, m=2$')
ax1.plot(t, h_iner_2m2, color='dodgerblue', alpha=0.8, label=r'$\ell=2, m=-2$')
ax1.plot(t, h_iner_4m4, color='green',label=r'$\ell=4, m=-4$')
ax1.hlines(y=0., xmin=100, xmax=200, color='darkorange')
ax1.hlines(y=0., xmin=100, xmax=200, color='dodgerblue')
ax1.hlines(y=0., xmin=100, xmax=200, color='green')
ax1.set_ylabel(r'Re$(h_{\ell,m})\cdot r/M$')
ax1.set_xlabel(r'$t/$M')
ax1.legend()
ax1.set_xlim(-4300,200)
ax1.set_ylim(-0.599999,0.57)

# inset plot
axins.plot(t, h_iner_22, color='darkorange', label=r'$\ell=2, m=2$')
axins.plot(t, h_iner_2m2, color='dodgerblue', alpha=0.8, label=r'$\ell=2, m=-2$')
axins.plot(t, h_iner_4m4, color='green',label=r'$\ell=4, m=-4$')
x1, x2, y1, y2 = -100, 75, -.48, .54   # specify the limits
axins.set_xlim(x1, x2) 
axins.set_ylim(y1, y2) 
axins.set_xticklabels([])
axins.set_yticklabels([])
axins.set_xticks([])
axins.set_yticks([])
mark_inset(ax1, axins, loc1=4, loc2=1, fc="none", ec="0.5")

# save and show
fig.tight_layout()
#fig.savefig('modes_superkick.pdf')
plt.show()
```
