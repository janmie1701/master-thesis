```python
# plus and cross polarisation
def h(A, f, t):
    return A * np.sin(1 * np.pi * f * t)
```

```python
# main plot function to show and simulate GW effect
def simulate_particles(radius, num_particles, amplitude, frequency, time):
    angles = np.linspace(0, 2*np.pi, num_particles, endpoint=False)
    x_positions = radius * np.cos(angles)
    y_positions = radius * np.sin(angles)

    fig = plt.figure(figsize=(20, 5))

    for idx, t in enumerate(time):

        # Bewege die Teilchen entsprechend der Gravitationswelle für Plus-Polarisierung
        x_displacement_plus = h(amplitude, frequency, t) * np.sin(angles+np.pi/2)
        y_displacement_plus = h(amplitude, frequency, t) * np.cos(angles+np.pi/2)

        # Bewege die Teilchen entsprechend der Gravitationswelle für Kreuz-Polarisierung
        x_displacement_cross = h(amplitude, frequency, t) * np.sin(angles)
        y_displacement_cross = h(amplitude, frequency, t) * np.cos(angles)

        # Aktualisiere die Positionen der Teilchen
        x_positions_plus = x_positions + x_displacement_plus
        y_positions_plus = y_positions + y_displacement_plus

        x_positions_cross = x_positions + x_displacement_cross
        y_positions_cross = y_positions + y_displacement_cross

        # Visualisiere die Bewegung der Teilchen für Plus-Polarisierung
        #if idx == 3:
        #    plt.arrow(0.2*radius, 10*radius, 0.6*radius, 0, head_width=0.15*radius, head_length=0.2*radius, 
        #              fc='black', ec='black')
        if idx == 1:
            plt.text(-2.1*radius, 3*radius, 'time $\longrightarrow$', ha='center', fontsize=25)
            plt.text(-3*radius, 5.5*radius, r'$h_+$', ha='center', fontsize=25)
            plt.text(-3*radius, 0*radius, r'$h_{\times}$', ha='center', fontsize=25)
            

        
        # Visualisiere die Bewegung der Teilchen für Plus-Polarisierung
        plt.subplot(2, len(time), idx+1)
        plt.scatter(x_positions_plus, y_positions_plus, color='b', s=10, zorder=1)
        plt.xlim(-2*radius, 2*radius)
        plt.ylim(-2*radius, 2*radius)
        plt.gca().set_aspect('equal', adjustable='box')
        plt.axis('off')
        
        
        # Zeichne das Koordinatensystem 
        plt.annotate("", xy=(2*radius, 0), xycoords="data", xytext=(-0.05*radius, 0*radius), 
                     textcoords="data", arrowprops=dict(arrowstyle="->", linewidth=1.5), zorder=0)
        plt.annotate("", xy=(0, 2*radius), xycoords="data", xytext=(0*radius, -0.05*radius), 
             textcoords="data", arrowprops=dict(arrowstyle="->", linewidth=1.5), zorder=0)
        plt.text(-0.5*radius, 1.75*radius, "$y$", ha='center')
        plt.text(2*radius, -0.75*radius, "$x$", ha='center')


        # Visualisiere die Bewegung der Teilchen für Kreuz-Polarisierung
        plt.subplot(2, len(time), len(time)+idx+1)
        plt.scatter(x_positions_cross, y_positions_cross, color='r', s=10, zorder=1)
        plt.xlim(-2*radius, 2*radius)
        plt.ylim(-2*radius, 2*radius)
        plt.gca().set_aspect('equal', adjustable='box')
        plt.axis('off')
        
        
        # Zeichne das Koordinatensystem 
        plt.annotate("", xy=(2*radius, 0), xycoords="data", xytext=(-0.05*radius, 0*radius), 
                     textcoords="data", arrowprops=dict(arrowstyle="->", linewidth=1.5),  zorder=0)
        plt.annotate("", xy=(0, 2*radius), xycoords="data", xytext=(0*radius, -0.05*radius), 
             textcoords="data", arrowprops=dict(arrowstyle="->", linewidth=1.5) , zorder=0)
        plt.text(-0.5*radius, 1.75*radius, "$y$", ha='center')
        plt.text(2*radius, -0.75*radius, "$x$", ha='center')

    # save and show
    #fig.savefig('Polarization_Rings.pdf')
    plt.show()

# parameter for simulation
ring_radius = 1.0
num_particles = 20
grav_wave_amplitude = 0.82
grav_wave_frequency = 1.0
simulation_time = np.linspace(0, 2, 9)  

# Führe die Simulation durch und visualisiere die Ergebnisse
simulate_particles(ring_radius, num_particles, grav_wave_amplitude, grav_wave_frequency, simulation_time)
```
